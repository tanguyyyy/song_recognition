```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.io.wavfile import read, write
from scipy.signal import spectrogram
```

L'idée de base de l'algorithme est d'associer à chaque morceau de musique une empreinte calculée à partir de son spectrogramme, puis de créer a priori une base de données contenant les empreintes calculées pour plusieurs millions de morceaux. Pour reconnaître un extrait musical, l'algorithme procède ensuite en calculant de manière analogue une empreinte pour cet extrait, puis en cherchant une correspondance entre l'empreinte de cet extrait et une des empreintes de la base de donnée.

Spectrogramme

Le spectrogramme permet d'identifier les structures spatio-fréquentielles (les "notes") présentes dans un morceau de musique et constitue donc un excellent candidat pour le calcul de l'empreinte du morceau.

1. Calculer la transformée de Fourier fenêtrée d'un des signaux musicaux. Reconstruire le signal à partir de 10%, puis 5% des coefficients de la transformée de Fourier de plus grande amplitude. Commenter.

2. Bruiter l'un des signaux musicaux en lui ajoutant un bruit blanc gaussien, et appliquer la méthode de reconstruction non-linéaire détaillée dans la question précédente. Enregistrer le signal bruité et le signal reconstruit. Commenter le résultat obtenu.


```python
# Reconstruction du signal à partir de sa TF fenêtrée
from scipy.signal import stft, istft

# Chargement et affichage du signal
song = 'Madonna - Music'
filename = './samples/' + song + '.wav'
fs, s = read(filename)
s = s[:2048064]

f, t, Sxx = stft(s, fs=fs, window='hann', nperseg=256, noverlap=64)
plt.pcolormesh(t, f/1e3, np.abs(Sxx), shading='gouraud')
plt.xlabel('Time [s]')
plt.ylabel('Frequency [kHz]')
plt.show()

# Seuillage des coefficients de plus petite amplitude et reconstruction
coefs = np.sort(np.abs(Sxx).ravel())
threshold = coefs[int(0.9*coefs.size)]
print(threshold)
mask = (np.abs(Sxx) > threshold)
Sxx = Sxx*mask
_, srec = istft(Sxx, fs=fs, window='hann', nperseg=256, noverlap=128)
print('Euclidean distance:' + np.str(np.linalg.norm(s - srec)))

# Save compressed signal
write(song + ' - compressed.wav',fs, srec)

```
